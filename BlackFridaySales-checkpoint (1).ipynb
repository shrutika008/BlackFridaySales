# Cleaning and preparing the data for model training

# Z-test

### H0: There is no significane difference between mean sale of two product category(1 and 2)
### H1: There is  significane difference between mean sale of two product category(1 and 2)

```python
from scipy import stats
from statsmodels.stats import weightstats as stests
```

```python
prod_cat1 = df['Product_Category_1']
```

```python
prod_cat1
```

```python
cat1_mean = np.mean(prod_cat1)
print(cat1_mean)
```

```python
prod_cat2 = df['Product_Category_2']
```

```python
prod_cat2
```

```python
cat2_mean = np.mean(prod_cat2)
print(cat2_mean)
```

```python
a = np.array([cat1_mean,cat2_mean])
a
```

```python
ztest, pvalue = stests.ztest(a)
ztest, pvalue
```

### p-vale < 0.05, therefor we reject H0

### H0: There is no significane difference between mean sale of two product category(1 and 3)
### H1: There is  significane difference between mean sale of two product category(1 and 3)

```python
prod_cat3 = df['Product_Category_3']
```

```python
prod_cat3
```

```python
cat3_mean = np.mean(prod_cat3)
print(cat3_mean)
```

```python
b = np.array([cat1_mean,cat3_mean])
b
```

```python
ztest, pvalue = stests.ztest(b)
ztest, pvalue
```

### p-vale < 0.05, therefor we reject H0
